```python
from bokeh.plotting import figure 
from bokeh.io import output_notebook, show
from functools import lru_cache
from os.path import dirname, join

import pandas as pd
import os

from bokeh.io import curdoc
from bokeh.layouts import column, row
from bokeh.models import ColumnDataSource, PreText, Select
from bokeh.plotting import figure
from bokeh.embed import json_item
import json

DATA_DIR = join(os.getcwd(), 'data/datasets/meteorologico')
print(os.listdir(DATA_DIR))
output_notebook()
```

```python
# Esta funcion es para cargar el dataset
# por ahora esta tomada del ejemplo de stocks o sea de un csv
# en algun momento hay que hacer algo generico para leer cualquier fuente de datos
# estas tablas tienen información diaría por lo que nosotros quizas tengamos que cambiar el tratamiendo de esa columna para que considere mas nivel de granularidad
# TODO : para que es esto ? @lru_cache() 
def load_sensor_data(sensor='sensor1'):
    fname = join(DATA_DIR, 'daily_%s.csv' % sensor)
    data = pd.read_csv(fname, header=None, parse_dates=['date'],
                       names=['date', 'value']) # TODO: de aca solo usamos c (la penultima columna) y la primera (date)
    data = data.set_index('date')
    data['date_col'] = pd.to_datetime(data.index)
    #return pd.DataFrame({'value': data['value']}) # TODO : vamos a necesitar tener la diferencia con el datapoint anterior? 
    return data

# Generar un grafico de linea
def build_line_plot(data,sensor_name,width=1000,height=300,tools = 'pan,wheel_zoom,xbox_select,reset'):
    source_static = ColumnDataSource(data=dict(date=[], value=[]))
    source_static.data = data
    line_plot = figure(plot_width=width, plot_height=height, x_axis_type='datetime',title=sensor_name)
    #line_plot.line('date', 'value', source=source_static)
    line_plot.line(data['date_col'], data['value'])
    stats = PreText(text='')
    stats.text = str(data.describe())
    #TODO parte de stats para ver si podemos hacerla reactiva a la seleccion
    layout = row(stats,line_plot)
    return layout
```

```python
sensor = 'sensor2'
sdata = load_sensor_data(sensor)
sline = build_line_plot(sdata,sensor)
show(sline)
```
